```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = 10000
x_behavior = np.random.beta(5,2, N)
x_target = np.random.beta(15,2, N)

#sns.distplot(x_behavior, label='behavior')
sns.distplot(x_target, label='target')
plt.legend()

q_0 = np.quantile(x_behavior, 0.05)
q_1 = np.quantile(x_behavior, 0.95)
```

```python
y = np.linspace(0, 1, 1000)
s0 = lambda y: q_0 - y
s1 = lambda y: y - q_1
s = lambda y: np.maximum(s0(y), s1(y))


data = x_behavior
sns.distplot(s(data), label='max(s0, s1)')
sns.distplot(s0(data), label='s0')
sns.distplot(s1(data), label='s1')

plt.legend()

q9_s = np.quantile(s(data), 0.9)
q9_s0 = np.quantile(s0(data), 0.9)
q9_s1 = np.quantile(s1(data), 0.9)
print(f'{q9_s} - {q9_s0} - {q9_s1}')


plt.figure()
plt.plot(y, s0(y), label='$s_0(x,y)$')
plt.plot(y, s1(y), label='$s_1(x,y)$')
plt.plot(y, s(y), label='$s(x,y)$')
plt.plot(y, q9_s * np.ones(len(y)),'k--', label='0.9-Quantile $s(x,y)$')
plt.plot(y, q9_s0 * np.ones(len(y)),'k--', label='0.9-Quantile $s_0(x,y)$')
plt.plot(y, q9_s1 * np.ones(len(y)),'k--', label='0.9-Quantile $s_1(x,y)$')

plt.annotate( '0.9-Quantile $s_1(x,y)$', (0.1, q9_s1 + 0.01))
plt.annotate( '0.9-Quantile $s_0(x,y)$', (0.1, q9_s0 + 0.01))
plt.annotate( '0.9-Quantile $s(x,y)$', (0.6, q9_s + 0.01))
plt.legend()
plt.show()

```
